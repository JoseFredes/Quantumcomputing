```python

from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram
```

```python
#front
circ = QuantumCircuit(2,2) #setea los qbits y losbits, es normal que sea la misma cantidad de bits qu de qubits
circ.h(0) #puerta H
circ.x(1) #puerta x 
circ.z(1) #puerta Z
circ.measure((0,1),(0,1)) #a que qubits vamos a medir y el otro es a que bits vamosa  referenciar esos qbits
circ.draw() #dibuja el cirtcuito
```

```python
#backend
backend = Aer.get_backend("qasm_simulator")
job = execute(circ, backend, shots = 1024)
result = job.result()
count = result.get_counts(circ)
plot_histogram(count)
```
